```python
titledata = pd.read_sql("SELECT t.title, s.salary FROM employees e LEFT JOIN titles t ON e.emp_title_id = t.title_id LEFT JOIN salaries s ON e.emp_no=s.emp_no", connection)
```

```python
grouped = titledata.groupby('title').mean('salary')
```

```python
grouped.rename(columns={'salary':'avg. salary'},inplace=True)
grouped.reset_index(inplace=True)
grouped.sort_values('avg. salary',ascending=False,inplace=True)
```
